### Combining Org Data
Combining organizations that both sources have rated

```python
import json, pandas as pd
from pandas.io.json import json_normalize

dfAllSides = pd.read_json('CleanedAllSides.json', orient='index')

#dfAllSides

dfMBFC = pd.read_json('CleanedMBFC.json', orient='index')

#dfMBFC

dfMerged = pd.merge(dfAllSides, dfMBFC, on='name')
dfMerged = dfMerged.drop_duplicates()
dfMerged
```
